```python
def check_direction(value_a: int, value_b: int):
    #print('value_a: ' + str(value_a) + ' value_b: ' + str(value_b))
    if int(value_a) < int(value_b):
        #print('ascending')
        return 'ascending'
    elif int(value_a) > int(value_b):
        #print('descending')
        return 'descending'
    else:
        #print('error')
        return 'error'

def verify_stable_line_direction(values: list):
    line_ok = True
    direction = check_direction(values[0], values[1])
    for i in range(len(values) - 1):
        direction_new = check_direction(values[i], values[i + 1])
        if direction_new != direction:
            line_ok = False
            break
    #print('line direction is: ' + str(line_ok))
    return line_ok

def verify_stepsize(values: list):
    line_ok = True
    for i in range(len(values) - 1):
        stepsize = int(values[i])-int(values[i+1])
        stepsize = abs(stepsize)
        if stepsize > 3 or stepsize == 0:
            line_ok = False
            break
    #print('stepsize is: ' + str(line_ok))
    return line_ok
```

```python
def is_safe(row):
    inc = [row[i + 1] - row[i] for i in range(len(row) - 1)]
    if set(inc) <= {1, 2, 3} or set(inc) <= {-1, -2, -3}:
        return True
    return False

data = [[int(y) for y in x.split(' ')] for x in open('input2.txt').read().split('\n')]

safe_count = sum([is_safe(row) for row in data])
print(safe_count)

safe_count = sum([any([is_safe(row[:i] + row[i + 1:]) for i in range(len(row))]) for row in data])
print(safe_count)
```
